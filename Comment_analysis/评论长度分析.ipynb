```python
import pandas as pd
```

```python
filename="spider_comments_34000.csv"
data=pd.read_csv(filename,encoding="utf-8")
print("处理前数据条数:",data.shape[0])
data
```

```python
# 进行去重操作
data.drop_duplicates(['comment'],inplace=True)
print("去重后数据条数:",data.shape[0])
data
```

```python
# 去除含空值行
data.dropna(axis=0,how='any',inplace=True)
print("去除含空值行后数据条数:",data.shape[0])
data
```

```python
# 添加评论长度
lenthlist=[]
for index in data.index:
    lenth=len(data.loc[index].values[4])
    lenthlist.append(lenth)
data['contentlenth']=lenthlist
data
```

```python
import matplotlib.pyplot as plt
    
plt.hist(x = data.contentlenth,bins = 30,range=(0,200),color = 'steelblue')
plt.show()
```

```python
def classify_byscore(goodscore,badscore,df):
    goodlist=df[df["score"]>=goodscore]
    badlist=df[df["score"]<=badscore]
    return goodlist,badlist
def classify_bysentiment(goodsentiment,badsentiment,df):
    goodlist=df[df["sentiment"]>=goodsentiment]
    badlist=df[df["sentiment"]<=badsentiment]
    return goodlist,badlist
```

```python
# 选定分类方法（星级或情感分数）以及评判线
goodlist,badlist=classify_byscore(4,3,data)
```

```python
plt.hist(x = goodlist.contentlenth,bins = 80,range=(0,300),color = 'steelblue')
plt.ylabel("num")
plt.xlabel("comment_lenth")
plt.title("good_comment")
plt.show()
```

```python
plt.hist(x = badlist.contentlenth,bins = 80,range=(0,300),color = 'steelblue')
plt.ylabel("num")
plt.xlabel("comment_lenth")
plt.title("bad_comment")
plt.show()
```
