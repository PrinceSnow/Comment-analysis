```python
import pandas as pd
```

```python
infoname="spider_product_info_34000.csv"
info=pd.read_csv(infoname,encoding="utf-8")
info
```

```python
def classify_by_productid(df,info,suffix):
    for index,product in info.iterrows():
        product_id=product['product_id']
        outputName=product['brand']+product['model']
        df[df['product_id']==product_id].to_csv(outputName+suffix+"_comment.csv",header=True,index=False)
```

```python
filename="service_comment.csv"
suffix="service"
df=pd.read_csv(filename,encoding="utf-8")
classify_by_productid(df,info,suffix)
```
