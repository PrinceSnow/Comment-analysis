```python
service_list=[]
appearance_list=[]
noise_list=[]
consumption_list=[]
speed_list=[]
effect_list=[]
for idx,line in data.iterrows():
    service=re.findall(r"^安装服务：(.+?)\n", line['comment']+"\n", re.M)
    if not len(service)==0:
        service_list.append((line['product_id'],line['score'],line['comment_time'],service[0]))
        
    appearance=re.findall(r"^外形外观：(.+?)\n", line['comment']+"\n", re.M)
    if not len(appearance)==0:
        appearance_list.append((line['product_id'],line['score'],line['comment_time'],appearance[0]))
        
    noise=re.findall(r"^噪音大小：(.+?)\n", line['comment']+"\n", re.M)
    if not len(noise)==0:
        noise_list.append((line['product_id'],line['score'],line['comment_time'],noise[0]))
        
    consumption=re.findall(r"^耗能情况：(.+?)\n", line['comment']+"\n", re.M)
    if not len(consumption)==0:
        consumption_list.append((line['product_id'],line['score'],line['comment_time'],consumption[0]))
        
    speed=re.findall(r"^出水速度：(.+?)\n", line['comment']+"\n", re.M)
    if not len(speed)==0:
        speed_list.append((line['product_id'],line['score'],line['comment_time'],speed[0]))
    
    effect=re.findall(r"^恒温效果：(.+?)\n", line['comment']+"\n", re.M)
    if not len(effect)==0:
        effect_list.append((line['product_id'],line['score'],line['comment_time'],effect[0]))
        
```

```python
service_df=pd.DataFrame(service_list)
appearance_df=pd.DataFrame(appearance_list)
noise_df=pd.DataFrame(noise_list)
consumption_df=pd.DataFrame(consumption_list)
speed_df=pd.DataFrame(speed_list)
effect_df=pd.DataFrame(effect_list)
```

```python
import pandas as pd
import operator
import os
import re


def judgeRepeat(L1,L2):
    return operator.eq(L1,L2)
def compressed(comment):
    # 初始化两个空列表L1与L2
    L1 = []
    L2 = []
    # 初始化机械压缩后的评论
    compressedComment = []
    # 循环评论列表的每一个字符进行判断
    for letter in comment:
        # 如果L1为空，直接将字符放入L1中
        if len(L1)==0:
            L1.append(letter)
        else:
            # L1不为空且L1的第一个字符和读入字符相同
            if L1[0]==letter:
                # 如果L2为空，放入L2中
                if len(L2)==0:
                    L2.append(letter)
                else:
                    # L2不为空，触发压缩判断，如果得到重复
                    if judgeRepeat(L1,L2):
                    # 清空L2，然后将这个单词加入到L2中
                        L2.clear()
                        L2.append(letter)
                    else:
                        # 将L1,L2的内容加入到压缩评论中，清空L1,L2，将读取字符加入到L1
                        compressedComment.extend(L1)
                        compressedComment.extend(L2)
                        L1.clear()
                        L2.clear()
                        L1.append(letter)
            # 当前读入的字符和第一个列表的首字符不相同
            else:
                # 触发压缩判断，如果得出重复，并且列表含的字符长度大于等于2
                if len(L2)>=2 and judgeRepeat(L1,L2):
                    # 进行压缩去除，清空两个列表，将读入字符加入到L1
                    compressedComment.extend(L1)
                    L1.clear()
                    L2.clear()
                    L1.append(letter)
                else:
                    # 如果L2为空，将读入字符加入到L1
                    if len(L2)==0:
                        L1.append(letter)
                    else:
                    # 如果L2不为空，将读入字符加入到L2
                        L2.append(letter)
    # 读完所有字符后，触发压缩判断，对L1，L2有意义部分进行比较，若得出重复，则进行压缩去除。
    if  judgeRepeat(L1,L2):
        compressedComment.extend(L1)
    else:
        compressedComment.extend(L1)
        compressedComment.extend(L2)
    return compressedComment
```

```python
def compress_comment(line):
    comment=line['comment']
    comment = re.sub("[\s+\.\!V,$%^*(+\"\"]+|[!！+-,.，：。？?、~@#$%......&*();`:]+","",comment)
    commList = list(comment)
    commList = compressed(commList) #压缩前缀重复
    commList = compressed(commList[::-1]) #压缩后缀重复
    commList = commList[::-1]#把已经反转的列表再转置回来
    #  如果压缩后的评论长度小于2，则删除
    if len(commList)<2:
        return ""
    return "".join(commList).replace("\\n","")
```

```python
def compress_total(df):
    df['comment']=df.apply(compress_comment,axis=1) 
    df.drop(df[df['comment']==""].index,inplace=True)
```

```python
compress_total(service_df)
compress_total(appearance_df)
compress_total(noise_df)
compress_total(consumption_df)
compress_total(speed_df)
compress_total(effect_df)
```
