```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename=""
appearance_df=pd.read_csv(filename+"appearance_comment.csv")
consumption_df=pd.read_csv(filename+"consumption_comment.csv")
effect_df=pd.read_csv(filename+"effect_comment.csv")
noise_df=pd.read_csv(filename+"noise_comment.csv")
service_df=pd.read_csv(filename+"service_comment.csv")
speed_df=pd.read_csv(filename+"speed_comment.csv")
```

```python
def classify_byscore(goodscore,badscore,df):
    goodlist=df[df["score"]>=goodscore]
    badlist=df[df["score"]<=badscore]
    return [goodlist.shape[0],badlist.shape[0]]
```

```python
matrix=[]
matrix.append(classify_byscore(4,3,appearance_df))
matrix.append(classify_byscore(4,3,consumption_df))
matrix.append(classify_byscore(4,3,effect_df))
matrix.append(classify_byscore(4,3,noise_df))
matrix.append(classify_byscore(4,3,service_df))
matrix.append(classify_byscore(4,3,speed_df))
matrix=np.array(matrix)
```

```python
# df=pd.DataFrame(matrix,columns=['好评', '差评'],index=['外形外观', '耗能情况', '恒温效果','噪音大小','安装服务','出水速度'])
df=pd.DataFrame(matrix,columns=['good', 'bad'],index=['appearance', 'consumption', 'effect','noise','service','speed'])
df
```

```python
df.plot.bar(stacked=True)
plt.show()
```

```python
df2=pd.DataFrame(matrix.T,columns=['appearance', 'consumption', 'effect','noise','service','speed'],index=['good', 'bad'])
df2
```

```python
df2.plot.bar(stacked=True)
plt.show()
```
